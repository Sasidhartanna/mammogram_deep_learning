```python
## Import OpenCV
import numpy as np
import cv2
import matplotlib.pyplot as plt

%matplotlib inline
## Read the image
img = cv2.imread("mammo.pgm")
plt.imshow(img)
```

```python
img = cv2.imread("mammo.pgm",0)
plt.imshow(img)
```

```python
## Do the processing
i = 3  # Draw contour(i)
# gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY) # Convert to binary image
ret,thresh = cv2.threshold(img,30,255,0)

## Canny edge detectionto find threshold insead of using the above function(cv2.threshold())
# thresh=cv2.Canny(img,100,200)
plt.imshow(thresh)
```

```python
kernel = np.ones((20,20),np.uint8)
opening = cv2.morphologyEx(thresh, cv2.MORPH_OPEN, kernel)
plt.imshow(opening, cmap="gray")
```

```python
kernel = np.ones((40,40),np.uint8)
erosion = cv2.erode(opening,kernel,iterations = 1)
plt.imshow(erosion, cmap="gray")
```

```python
merged = cv2.bitwise_and(img, img , mask=erosion)
plt.imshow(merged, cmap="gray")
```
